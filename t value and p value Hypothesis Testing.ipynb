# Hypothesis-Testing-  1-Sample-1-Tail-Test-Salmonella-Outbreak-
# Using T test and p value

* Null Hypothesis Ho as Mean Salmonella <= 0.3 
* Alternate Hypothesis Ha as Mean Salmonella > 0.3
* As No direct code for 1-sample 1-tail ttest available with unknown SD and arrays of means. 
* Hence we find probability using 1-sample 2-tail ttest and divide it by 2 to get 1-tail ttest

## Way 1

```python
=0.3
array=pd.Series([0.593,0.142,0.329,0.691,0.231,0.793,0.519,0.392,0.418])
```

```python
# 1-sample 2-tail ttest:   stats.ttest_1samp(array,)   
p_value= stats.ttest_1samp(array,0.3)
p_value 
```

```python
# Two tail probability
p_2t=p_value[1]
p_2t 
```

```python
# 1-tail probability
p_1t=p_2t/2   
p_1t    
```

```python
if p_1t < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis \n Conclusion: \n Mean Salmonella > 0.3\n Ice-cream factory responsible for people illness.")
else:
    print("we are accepting null hypothesis \n Conclusion: \n Mean Salmonella < 0.3 \n Ice-cream factory is not responsible for people illness.")
```

# Hypothesis-Testing-  1-Sample-1-Tail-Test On Supermarket
# Using T test and p value

* Population Parameters: Mean=120 
* Sample Parameters: n=80, Mean=130, SD=40, df=80-1= 79
* Null Hypothesis Ho : Avg spending <= $120 
* Alternate Hypothesis Ha : Avg spending > $120.


```python
# Find t_value = (s_mean - p_mean)/(s_SD/sqrt(n))
t_value=(130-120)/(40/np.sqrt(80))
t_value
#Or (130-120)/(40/pow(80,0.5))
```

```python
(130-120)/(40/pow(80,0.5))
```

```python
#for 1-sample 1-tail ttest   p_value=1-stats.t.cdf(t_value,dof=79)
# p-value for 2-sided test    p_value= 2*(1 - t.cdf(abs(t_stat), dof))
```

```python
# Find Probability for 1-sample 1-tail ttest
p_value=1-stats.t.cdf(t_value,df=79)
p_value
#Or stats.t.sf(2.23,df = 79)
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis \n Conclusion:\n We can go ahead and launch Loyality program.")
else:
    print("we are accepting null hypothesis \n Conclusion: \n Loyalaty Program should not be launch")
```

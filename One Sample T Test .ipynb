# Import required libraries

```python
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, ttest_rel
```

* Null Hypothesis H0: mu = 20 , Potato yeild for all potato is same as standard potato yield
* Alternate Hypothesis H1: mu > 20 , Potato yeild for other potatos can be greater than standard potato yield
* Population mean = 20

#### Create a sample

```python
x =  [21.5, 24.5, 18.5, 17.2, 14.5, 23.2, 22.1, 20.5, 19.4, 18.1, 24.1, 18.5]
```

```python
tscore, pvalue = ttest_1samp(x, popmean=20, alternative='two-sided')
```

```python
print("t Statistic: ", tscore,"\n"
      "P Value: ",pvalue)  
```

```python
# The p value, which we calculated is for 2 tailed test
# Our probleem statement is about 1 tailed test
# So we need to divide it by 2 to get one tailed test p value

p_value = pvalue/2
p_value
```

```python
if p_value < 0.05:    # alpha value is 0.05 or 5%
    print(" we are rejecting null hypothesis")
else:
    print("We do not rejecting the Null Hypothesis anyway. \n Conclusion: \n sample mean is not differs from the an expected (population) mean")
```
